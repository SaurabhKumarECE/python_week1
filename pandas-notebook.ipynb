```python
# Day 3
import pandas as pd

df = pd.read_csv('/kaggle/input/datasets/brendan45774/test-file/tested.csv')

df.head(10)

```

```python
df.info()
```

```python
# Select columns
df[['Age', 'Sex', 'Fare']]
```

```python
# Filter rows
df[df['Age'] > 30]
df[df['Survived'] == 1]
```

```python
print("Shape:", df.shape)
print("\nMissing Values:\n", df.isnull().sum())
print("\nDuplicates:", df.duplicated().sum())
print("\nData Types:\n", df.dtypes)
```

### Before Cleaning Analysis

- The dataset contains 418 rows and 12 columns.
- Missing values are present in:
  - Age column (86 missing values)
  - Fare column (1 missing value)
  - Cabin column (327 missing values, which is very high)
- No duplicate records are found in the dataset.
- Data types include both numerical and categorical values.
- The Cabin column has excessive missing values and may not be useful for analysis.

```python
# 1. Handle missing values

# Fill Age with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill Fare with median
df['Fare'].fillna(df['Fare'].median(), inplace=True)

# Drop Cabin column (too many missing values)
df.drop(columns=['Cabin'], inplace=True)

```

```python
# 2. Convert categorical to numeric

df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
# 3. Convert data types

df['Age'] = df['Age'].astype(int)

```

```python
# 4. Feature Engineering

df['Family_Size'] = df['SibSp'] + df['Parch']
```

```python
# After cleaning
print("Shape:", df.shape)
print("\nMissing Values:\n", df.isnull().sum())
print("\nDuplicates:", df.duplicated().sum())
print("\nData Types:\n", df.dtypes)
```

### After Cleaning Analysis

- Missing values in Age and Fare have been successfully filled using median values.
- The Cabin column has been removed due to excessive missing data.
- All columns now have no missing values.
- The dataset contains only relevant and clean data.
- Categorical values like Sex have been converted into numerical format for easier analysis.
- A new feature "Family_Size" has been created to enhance analysis.
