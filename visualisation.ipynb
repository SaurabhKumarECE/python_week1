```python
import pandas as pd

df = pd.read_csv('/kaggle/input/datasets/brendan45774/test-file/tested.csv')
df.head()
```

### Day 5

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

Insight:

Females had a significantly higher survival rate than males.

```python
df['Age'].plot(kind='hist', bins=20)
plt.title("Age Distribution")
plt.show()
```

Insight:

Most passengers were between 20–40 years old.

```python
sns.histplot(data=df, x='Age', hue='Survived', bins=20)
plt.title("Age vs Survival")
plt.show()
```

```python
df['Fare'].plot(kind='hist', bins=30)
plt.title("Fare Distribution")
plt.show()
```

Insight:

Most passengers paid lower fares, with few high-paying outliers.

```python
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title("Fare vs Survival")
plt.show()
```

Insight:

Higher fare passengers had better survival chances

```python
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title("Embarked vs Survival")
plt.show()
```

Insight:

Survival rates vary depending on boarding location.

```python
sns.countplot(x='Family_Size', hue='Survived', data=df)
plt.title("Family Size vs Survival")
plt.show()
```

Insight:

Small families had better survival chances than large ones.

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

Insight:

Survival is positively correlated with Fare and negatively with Passenger Class.

```python
survival_rate = df['Survived'].value_counts(normalize=True) * 100
survival_rate.plot(kind='bar')
plt.title("Survival Percentage")
plt.show()
```
