```python
import numpy as np

# 1. Create 1D array
arr1 = np.array([1, 2, 3, 4, 5])
print(arr1)
```

```python
# 2. Create 2D array
arr2 = np.array([[1, 2], [3, 4]])
print(arr2)
```

```python
# 3. Matrix Addition
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(A + B)
```

```python

# 4. Matrix Multiplication
print(np.dot(A, B))

```

```python
# 5. Mean & Std
print("Mean:", np.mean(arr1))
print("Std:", np.std(arr1))
```

```python
# 1. Random matrix
rand_mat = np.random.rand(3,3)
print(rand_mat)
```

```python
# 2. Column Sum
print(np.sum(rand_mat, axis=0))

```

```python
# 3. Row Mean
print(np.mean(rand_mat, axis=1))
```

```python
# 4. Standard deviation
print(np.std(rand_mat))
```

```python
# 5. Transpose
print(rand_mat.T)
```
